### Manualmente, são colocadas chapas de metal no dispositivo e, mediante um cilindro pneumático, a chapa é fixada. Com outros dois cilindros, a chapa é dobrada e um outro efetua a estampagem.

```python
# Definição do sistema [Pressão em MPa e fluxo de ar, em l/min]:
circuit = PNEUMATIC(1.2, 5)

# Definição dos cilindros pneumáticos [nome, curso em mm, diâmetro maior em mm, diâmetro menor em mm e posição inicial em mm]:
circuit.addActuator('A', 100, 20, 8, 0)
circuit.addActuator('B', 100, 20, 8, 0)
circuit.addActuator('C', 100, 20, 8, 0)

# Sequência de comandos:
circuit.addComand('A+') # Prende a peça
circuit.addComand('B+') # Dobra
circuit.addComand('C+') # Estampa/dobra
circuit.addComand('C-') # Recolhe estampa
circuit.addComand('B-') # Recolhe cilindro da dobra
circuit.addComand('A-') # Solta a peça

# Executa a simulação:
circuit.simulate()
```
