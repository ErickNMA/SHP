```python
from pneumatic import *
```

```python
# Definio do sistema [Presso em MPa e fluxo de ar, em l/min]:
circuit = PNEUMATIC(0.6, 1)

# Definio dos cilindros pneumticos [nome, curso em mm, dimetro maior em mm, dimetro menor em mm e posio inicial em mm]:
circuit.addActuator('A', 100, 20, 20, 0)
circuit.addActuator('B', 100, 20, 20, 0)
circuit.addActuator('C', 100, 20, 20, 0)

# Sequncia de comandos:
circuit.addComand('A+')
circuit.addComand('B+')
circuit.addComand('C+')
circuit.addComand('C-')
circuit.addComand('B-')
circuit.addComand('A-')

# Executa a simulao:
circuit.simulate()
```
