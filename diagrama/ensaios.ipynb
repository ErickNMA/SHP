```python
from pneumatic import *
```

```python
# Definição do sistema [Pressão em MPa e fluxo de ar, em l/min]:
circuit = PNEUMATIC(0.6, 1)

# Definição dos cilindros pneumáticos [nome, curso em mm, diâmetro maior em mm, diâmetro menor em mm e posição inicial em mm]:
circuit.addActuator('A', 100, 20, 20, 0)
circuit.addActuator('B', 100, 20, 20, 0)
circuit.addActuator('C', 100, 20, 20, 0)

# Sequência de comandos:
circuit.addComand('A+')
circuit.addComand('B+')
circuit.addComand('C+')
circuit.addComand('C-')
circuit.addComand('B-')
circuit.addComand('A-')

# Executa a simulação:
circuit.simulate()
```
