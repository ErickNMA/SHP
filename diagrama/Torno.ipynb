### Para o acabamento de buchas, internamente, estas devem ser conduzidas por intermédio de um carrinho que avança, mediante o cilindro (1.0). Quando o carrinho chega na posição final dianteira, o cilindro (2.0) empurra a bucha entre as castanhas da placa. O cilindro (3.0) aciona a placa e a bucha é fixada. A unidade de avanço hidropneumática (4.0) movimenta o carro porta-ferramenta, executando o trabalho. A peça é solta e retirada manualmente.

```python
# Definição do sistema [Pressão em MPa e fluxo de ar, em l/min]:
circuit = PNEUMATIC(0.6, 5)

# Definição dos cilindros pneumáticos [nome, curso em mm, diâmetro maior em mm, diâmetro menor em mm e posição inicial em mm]:
circuit.addActuator('A', 100, 40, 10, 0)
circuit.addActuator('B', 50, 20, 8, 0)
circuit.addActuator('C', 100, 30, 8, 0)
circuit.addActuator('D', 200, 50, 30, 0)

# Sequência de comandos:
circuit.addComand('A+') # Avança o carrinho
circuit.addComand('B+') # Posiciona a bucha
circuit.addComand('C+') # Aciona a castanha
circuit.addComand('D+') # Movimenta o carro porta-ferramenta (executa o trabalho)
circuit.addComand('D-') # Retorna o carrinho principal
circuit.addComand('C-') # Libera a castanha
circuit.addComand('B-') # Retorna a bucha
circuit.addComand('A-') # Retorna o carrinho

# Executa a simulação:
circuit.simulate()
```
